```python
import tensorflow as tf
import numpy as np

batch = 3
feature = 10 
# the vectorization size of the word / input
#(it depends on how the word has been Embedded / Vectorized for us to know the feature size...)

hidden_size = 20 
# memory size
#you decide this.. no real way to know it...

timesteps = 2 
# sequence length (or no. of words)

inputs = np.random.randn(batch, timesteps, feature)

lstm = tf.keras.layers.LSTM(hidden_size)

h_out = lstm(inputs)
```

```python
print('Output size(batch, hidden_size) = ', h_out.shape)
```
